```python
#Importamos las librerias para el procesamiento de datos
import numpy as np
import pandas as pd
import shutil as sh
from sklearn.model_selection import train_test_split
```

```python
# Lectura del dataset
data = pd.read_csv('PRSA_Data_Guanyuan_20130301-20170228.csv')
data
```

```python
# Supongamos que quieres ver cuántas clases distintas hay en la columna 'clase'
num_clases = data['RAIN'].nunique()

# Imprime el número de clases distintas
print("Número de clases distintas:", num_clases)
```

```python
#Leemos los datos y la cantidad de los mismos
print('DATOS')
print(data)
print('\nCANTIDAD')
print(data.shape)
```

```python
#Leemos el tipo de datos que tiene el dataset y vemos los datos faltantes
print('INFORMACION DE TIPO DE DATOS')
data.info()
print('\nDATOS VACIOS')
print(pd.isnull(data).sum())
```

### PROCESAMIENTO DE DATOS

```python
# Para la imputación por la media
data['TEMP'].fillna(data['TEMP'].mean(), inplace=True)
data['PRES'].fillna(data['PRES'].mean(), inplace=True)
data['DEWP'].fillna(data['DEWP'].mean(), inplace=True)
data['RAIN'].fillna(data['RAIN'].mean(), inplace=True)
data['WSPM'].fillna(data['WSPM'].mean(), inplace=True)
```

```python
# CO y O3 tienen muchos más valores faltantes por lo que consideramos la interpolación para mantener la coherencia temporal de los datos
data['CO'].interpolate(method='linear', inplace=True)
data['O3'].interpolate(method='linear', inplace=True)
```

```python
print('\nDATOS VACIOS DESPUES DE IMPUTACION')
print(pd.isnull(data).sum())
```

```python
# Imputación por interpolación para contaminantes
contaminantes = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
for col in contaminantes:
    data[col].interpolate(method='linear', inplace=True)

# Imputación con la moda para la dirección del viento
data['wd'].fillna(data['wd'].mode()[0], inplace=True)

# Verificación final de datos faltantes
print('\nDATOS VACIOS DESPUES DE LA SEGUNDA IMPUTACION')
print(pd.isnull(data).sum())
```

```python
# Dividir en conjunto de entrenamiento y prueba
train, test = train_test_split(data, test_size=0.2, random_state=42, stratify=None)

# Verificar las formas de los conjuntos resultantes
print("Forma de train:", train.shape)
print("Forma de test:", test.shape)

# Guardar los conjuntos de datos
pd.DataFrame(train).to_csv('train.csv', index=False)
pd.DataFrame(test).to_csv('test.csv', index=False)
```

```python
# Guardamos el dataset ya procesado
data.to_csv('PRSA_processed.csv', index=False)
```
